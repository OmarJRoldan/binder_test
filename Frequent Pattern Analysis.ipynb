```python
import pandas as pd
```

```python
path = 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/fpEg1.csv'
```

```python
df = pd.read_csv(path,
                 header = None)
df.head()
```

```python
ndf = df.applymap(str)
ndf
```

```python
trans = ndf.values.tolist()
trans
```

```python
#pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_ary = te.fit(trans).transform(trans)
te_ary
```

```python
te_df = pd.DataFrame(te_ary,
                    columns = te.columns_)
te_df
```

```python
te_df = te_df.drop('nan',
                  axis = 1)
te_df
```

```python
top10 = te_df.head(10)
ndf = top10.copy()
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
frequent_items = fpgrowth(ndf,
                         min_support = 0.6,
                         use_colnames = True)
frequent_items
```

```python
as_ruls = association_rules(frequent_items,
                            metric = 'lift',
                            min_threshold = 0.2
                           )
as_ruls
#Si lift es negativo, entonces no se llevan
#Confidence 
```

### Case 1: What are the top 5 combinations with maximun confidence?

```python
as_ruls.sort_values('confidence',
                    ascending = False).head(5)
```

### Case 2: I want only antecedents (m)

```python
as_ruls['antecedents']=={'m'}
```

```python
### Case 3: I want only antecedents (c,f) and consequents (m)
```
